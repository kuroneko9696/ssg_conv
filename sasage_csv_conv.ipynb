```
#@title ①設定の読み込み
import pandas as pd
#ささげ共通リストの対応表です。

order ={
"r01"	:	"Null",    	#'案件名＝管理番号'
"r02"	:	"Null",	    #'顧客コード'
"r03"	:	"Null",	    #'ブランドコード'
"r04"	:	"Null",	    #'担当PM'
"r05"	:	"Null",	    #'スタジオ名'
"r06"	:	"Null",	    #'管理番号＝案件名に対する備考'
"r07"	:	"JANコード",	#'JAN'
"r08"	:	"型番",	    #'仮品番'
"r09"	:	"メーカー品番",	#'メーカー品番'
"r10"	:	"商品名",	  #'商品名'
"r11"	:	"カラー名",	#'カラー及び柄名'
"r12"	:	"カラーコード",	#'カラー及び柄\nコード'
"r13"	:	"サイズ名",	#'サイズ名'
"r14"	:	"サイズコード",	#'サイズ\nコード'
"r15"	:	"ブランド名",	#'ブランド名'
"r16"	:	"区分（性別）",	#'区分（性別）'
"r17"	:	"メインカラー選択",	#'メインカラー\n指定'
"r18"	:	"Null",	    #'商品に関する備考'
"r19"	:	"加工指示",	    #'商管備考'　>>>加工指示の項目は使いたい
"r20"	:	"採寸有無",	#'採寸有無'
"r21"	:	"計測箇所\nリンク",	#'採寸指示\n詳細'
"r22"	:	"撮影有無",	#'撮影有無'
"r23"	:	"撮影指示",	#'撮影指示詳細'
"r24"	:	"Null",	    #'レタッチ有無'
"r25"	:	"原稿入力欄",	    #'原稿有無'
"r26"	:	"Null",	    #'原稿指示\n詳細'
"r27"	:	"Null",	    #'希望納品日'
"r28"	:	"着荷日",	  #'着荷予定日'
"r29"	:	"納期",	    #'納品予定日'
"r30"	:	"箱番号",	  #'箱番号'
"r31"	:	"Null",	    #'入庫日'
"r32"	:	"Null",	    #'入庫（商管）担当者'
"r33"	:	"Null",	    #'返却日'
}


#------------------------------------------------------------------------------------------
#@markdown GoogleDriveをマウント（アクセス可能な状態に）します。
print("あなたがお使いのGoodleDrive上で動作するようにします")
from google.colab import drive
drive.mount('/content/drive/')
%cd /content/drive


#------------------------------------------------------------------------------------------
#@markdown 依頼リスト(顧客から入手したエクセルファイル）を読み込みます。
#実行するフォルダまで移動する必要があります。
path = "/content/drive/MyDrive/converter/csv"

%cd /content/drive/MyDrive/converter/csv
%pwd
#tmpファイルを削除し、空のtmpフォルダを作成する（勝手に連番でファイルを作ってしまうため）
import os
tmppath = './tmp'
#print(os.path.exists(tmppath))
if os.path.exists(tmppath) == False:
  %mkdir tmp
elif os.path.exists(tmppath) == True:
  %rm -rf tmp
  %mkdir tmp

%cd tmp
%pwd
from google.colab import files


#------------------------------------------------------------------------------------------
print("顧客からの依頼ファイル（エクセルをアップロードします")
print("正しい三喜様フォーマット以外のファイルは読み込めません")

uploaded = files.upload()
file_name = list(uploaded.keys())[0]
order_name = file_name.split(".")[0]
directory = %pwd
myfile = directory+"/"+file_name
skiprows = 1 #1行飛ばす
df = pd.read_excel(myfile,skiprows=skiprows)
#df.columns
print("読み込みました")

#------------------------------------------------------------------------------------------
new_pd = pd.DataFrame()
for iii in order.keys():
  if order[iii] == "Null":
    #print("対応する列がありません")
    new_pd[iii] = ""
    #new_pd[iii].replace('', np.nan, inplace=True)

  else:
    #print(order[iii])
    targe = df.loc[:,order[iii]]
    new_pd = pd.concat([new_pd, targe], axis=1)



#カラム名を一括書き換え
new_pd.columns = ['案件名＝管理番号', '顧客コード', 'ブランドコード', '担当PM', 'スタジオ名', '管理番号＝案件名に対する備考',
       'JAN', '仮品番', 'メーカー品番', '商品名', 'カラー及び柄名', 'カラー及び柄コード', 'サイズ名',
       'サイズコード', 'ブランド名', '区分（性別）', 'メインカラー指定', '商品に関する備考', '商管備考',
       '採寸有無', '採寸指示詳細', '撮影有無', '撮影指示詳細', 'レタッチ有無', '原稿有無', '原稿指示詳細',
       '希望納品日', '着荷予定日', '納品予定日', '箱番号', '入庫日', '入庫（商管）担当者', '返却日'
       ]



#カラム名を一括書き換え
#new_pd.columns = ['案件名＝管理番号', '顧客コード', 'ブランドコード', '担当PM', 'スタジオ名', '管理番号＝案件名に対する備考',
#       'JAN', '仮品番', 'メーカー品番', '商品名', 'カラー及び柄名', 'カラー及び柄\n  コード', 'サイズ名',
#       'サイズ\n  コード', 'ブランド名', '区分（性別）', 'メインカラー\n  指定', '商品に関する備考', '商管備考',
#       '採寸有無', '採寸指示\n  詳細', '撮影有無', '撮影指示詳細', 'レタッチ有無', '原稿有無', '原稿指示\n  詳細',
#       '希望納品日', '着荷予定日', '納品予定日', '箱番号', '入庫日', '入庫（商管）担当者', '返却日'
#       ]


#new_pd.style.highlight_null(null_color='pink',subset = [ 'JAN',  'メーカー品番'])
#new_pd
print(file_name,"をデータフレームに読み込みました")


#------------------------------------------------------------------------------------------
import numpy as np

#案件名
#ankenmmei = "anken name"
new_pd['案件名＝管理番号'] =  new_pd['案件名＝管理番号'].fillna(order_name) #暫定的にエクセルのファイル名を依頼名に

#顧客コード
kokyakucode = "k-code"
new_pd['顧客コード'] =  new_pd['顧客コード'].fillna(kokyakucode)

#ブランドコード
brandcode = "b-code"
new_pd['ブランドコード'] =  new_pd['ブランドコード'].fillna(brandcode)

#PM
pm = "pm-code"
new_pd['担当PM'] =  new_pd['担当PM'].fillna(pm)

#スタジオ名
studioname = "studio-code"
new_pd['スタジオ名'] =  new_pd['スタジオ名'].fillna(studioname)

#●をbool型にする関数
def convertNan2text(colname):
  #print(colname)
  for i, r in new_pd.iterrows():
    #print("-----"+str(i)+"--------",r)
    data = new_pd.loc[i,colname]
    #print(data,type(data))
    if type(data) == str:
      #print("-str---")
      #print(data,type(data))
      #print(data)
      if data == "●":
        new_pd.loc[i,colname] = "true"
      elif data == "有り":
        new_pd.loc[i,colname] = "true"

      elif data != "●":
        new_pd.loc[i,colname] = "false"

    elif type(data) == float:
      #print("-float--")
      data = str(data)
      #print(data,type(data))
      #print("ppppp",data,type(data))
      new_pd.loc[i,colname] = "false"

#CSVで改行コードが変になる対策（""でくくる）
def encloseBr(colname):
  #print(colname)
  for i, r in new_pd.iterrows():
    #print("-----"+str(i)+"--------",r)
    data = new_pd.loc[i,colname]
    #print(data,type(data))
    if type(data) == str:
        new_pd.loc[i,colname] = data.replace("\n","")
    elif type(data) == float:
        pass


#改行対策
colname = "撮影指示詳細"
encloseBr(colname)

#フラグ系のカラムを処理
colname = "メインカラー指定"
convertNan2text(colname)
colname = "撮影有無"
convertNan2text(colname)
colname = "採寸有無"
convertNan2text(colname)
colname = "レタッチ有無"
convertNan2text(colname)
colname = "原稿有無"
convertNan2text(colname)

#Nullを強調表示
#new_pd.style.highlight_null(null_color='red')
#csvに出力する
import datetime
now = datetime.datetime.now()
now = kokyakucode+str(now.year)+"-"+str(now.month)+"-"+str(now.day)
#new_pd.to_csv("/content/drive/MyDrive/"+now+".csv", index = False)
new_pd.to_csv("/content/drive/MyDrive/converter/csv/base.csv", index = False)

print("csvを出力しました")
print("ささげ共通リストコンバーター")
print("https://docs.google.com/spreadsheets/d/1crFSuyIhY03mnpO7WXmCQX-V9mLqslsPc-3ce0Dhk9s/edit#gid=0")
```
